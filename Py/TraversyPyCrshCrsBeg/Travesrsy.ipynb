```python
##CLASSES##

class User:
    #Construcotr
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def greeting(self):
        return f'My name is {self.name} and my age is {self.age}'

#Init
tin = User('Tin',31)
print(type(tin),tin.name)

print(tin.greeting())


##Extending a class
class Customer(User):
    #Construcotr
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.balance = 0
    def set_balance(self,balance):
        self.balance = balance
    def greeting(self):
        #print(super(Customer.__bases__[0],self).greeting())
        #super(Customer.__bases__[0], self).greeting()   
        return f' you have {self.balance}$'
        
        
cust = Customer('Tin',31)
cust.set_balance(12),

print(type(cust),cust.balance)
        
kor = [tin,cust]
    
for x in kor:
    print(x.greeting())
```

```python
##LOOPS##

people = ['John','Paul','Sarah','Suzy']


#FOR

for person in people:
    print(f'Current person {person}')
    
#Break  (stops)
print('')

for person in people:
    if person == 'Sarah':
        break
    print(f'Current person {person}')

#Continue (skips)
print('')

for person in people:
    if person == 'Sarah':
        continue
    print(f'Current person {person}')
    
#CRange
print('')

for i in range(len(people)):
    print(people[i])

print('')

for i in range(0,11,2):
    print(f'Number : {i}')

#While
print('')


cnt = 0

while cnt <= 10:
    print(cnt)
    cnt+=1
    
```
