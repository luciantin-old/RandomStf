```python
import cv2
import numpy as np

input = cv2.imread('./images/daisy.jpg')

B,G,R = cv2.split(input)

zeros = np.zeros(input.shape[:2],dtype='uint8')

cv2.imshow('B',cv2.merge([B,zeros,zeros]))
cv2.imshow('G',cv2.merge([zeros,G,zeros]))
cv2.imshow('R',cv2.merge([zeros,zeros,R]))

cv2.waitKey()

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 

image = np.zeros((512,512,3), np.uint8) 

image_bw = np.zeros((512,512), np.uint8)

cv2.imshow('image',image)
cv2.imshow('bw',image_bw)

cv2.waitKey()
cv2.destroyAllWindows()

cv2.line(image,(0,0),(511,511),(255,127,0), 5)
cv2.imshow('blue line',image)

cv2.waitKey()
cv2.destroyAllWindows()

cv2.rectangle(image,(100,100),(300,250),(255,127,0), 5)
cv2.imshow('blue line',image)

cv2.waitKey()
cv2.destroyAllWindows()

cv2.circle(image,(350,350),100,(0,127,0), -1)
cv2.imshow('blue line',image)

cv2.waitKey()
cv2.destroyAllWindows()

pts = np.array( [[10,50],[400,50],[90,200],[50,100]],np.int32 )
pts = pts.reshape((-1,1,2))

cv2.polylines(image,pts,True,(0,127,0), 3)
cv2.imshow('blue line',image)

cv2.waitKey()
cv2.destroyAllWindows()
```
