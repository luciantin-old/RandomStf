```python
import cv2
import numpy as np

image = cv2.imread('./images/shapes.jfif')

origImage = image.copy()

image = cv2.bitwise_not(image)
image = cv2.GaussianBlur(image, (5,5), 0)
image = cv2.erode(image, np.ones((5,5),np.uint8), iterations = 2)
image = cv2.dilate(image, np.ones((5,5),np.uint8), iterations = 1)
image = cv2.bitwise_not(image)

cv2.imshow('',image)
cv2.waitKey()

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
ret, gray = cv2.threshold(gray,245,255,cv2.THRESH_BINARY)

gray = cv2.bitwise_not(gray)

gray = cv2.erode(gray, np.ones((5,5),np.uint8), iterations = 2)
gray = cv2.dilate(gray, np.ones((5,5),np.uint8), iterations = 1)

cv2.imshow('',gray)
cv2.waitKey()

edge = cv2.Canny(gray,10,200)
cv2.imshow('',edge)
cv2.waitKey()

#changes the image sent to the function
contours, hierarchy = cv2.findContours(edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.imshow('',edge)
cv2.waitKey()

print('Num of cont '+ str(len(contours)))

cv2.drawContours(origImage,contours,-1,(0,255,0),3)

cv2.imshow('',origImage)
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np

image = cv2.imread('./images/shapes.jfif')

origImage = image.copy()
origImage2 = image.copy()

cv2.imshow('',image)
cv2.waitKey()

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

cv2.imshow('',gray)
cv2.waitKey()

edge = cv2.Canny(gray,20,255)
cv2.imshow('',edge)
cv2.waitKey()

#changes the image sent to the function
contours, hierarchy = cv2.findContours(edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow('',edge)
cv2.waitKey()
# print(contours)

# for x in contours:
#     for y in x:
#         cv2.circle(origImage,(y[0][0],y[0][1]),20,(255,0,0), -1)

print('Num of cont '+ str(len(contours)))

# -1 da nacrta sve konture
# cv2.drawContours(origImage,contours,-1,(0,255,0),3)
# cv2.imshow('',origImage)

sortedContours = sorted(contours , key=cv2.contourArea, reverse=True)

for c in sortedContours :
    print(c)
    cv2.drawContours(origImage,[c],-1,(0,255,0),4)
    cv2.imshow('',origImage)
    cv2.waitKey()

for (i,c) in enumerate(sortedContours) :
    (x,y,w,h) = cv2.boundingRect(c)
    croppedImage = origImage2[y:y+h,x:x+w] 
    cv2.imshow('',croppedImage)
    cv2.imwrite("shape_"+str(i)+'.jpg', croppedImage)
    cv2.waitKey()
    
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np

image = cv2.imread('./images/drawingsShapes.bmp')
origImage = image.copy()
origImage2 = image.copy()

cv2.imshow('',image)
cv2.waitKey()

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# edge = cv2.Canny(gray,20,255)
edge = cv2.bitwise_not(gray)
# cv2.imshow('',edge)
# cv2.waitKey()
                                                #EXTERNAL LIST
contours, hierarchy = cv2.findContours(edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.imshow('',edge)
cv2.waitKey()

cv2.drawContours(origImage,contours,-1,(0,255,0),4)
cv2.imshow('',origImage)
cv2.waitKey()

for c in contours:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(origImage,(x,y), (x+w,y+h), (255,0,0),2)

cv2.imshow('',origImage)
cv2.waitKey()
    
for c in contours:
    accuracy = 0.03 * cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, accuracy, True)
    cv2.drawContours(origImage, [approx], 0, (0,0,255),2)
    
cv2.imshow('',origImage)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np 

image = cv2.imread('./images/daisy.jpg')

cv2.imshow('',image)
cv2.waitKey()

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# cv2.imshow('',gray)
# cv2.waitKey()


kernel = np.ones((5,5), np.uint8)
# erosion = cv2.erode(gray,kernel,iterations = 2)
# cv2.imshow('',erosion)
# cv2.waitKey()

dilation = cv2.dilate(gray,kernel, iterations = 1)
# cv2.imshow('',dilation)
# cv2.waitKey()

ret, thresh1 = cv2.threshold(dilation,30,255,cv2.THRESH_BINARY)
# cv2.imshow('',thresh1)
# cv2.waitKey()

contours, hierarchy = cv2.findContours(thresh1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(image,contours,-1,(0,255,0),4)
cv2.imshow('',image)
cv2.waitKey()

#sortiraj i uzmi najcevu samo
# n = len(contours)-1
# contours = sorted(contours, key=cv2.contourArea, reverse=False)[:n]

for c in contours:
    hull = cv2.convexHull(c)
    cv2.drawContours(image,[hull],0,(255,0,0),3)
    cv2.imshow('',image)
    cv2.waitKey()

cv2.waitKey()
cv2.destroyAllWindows()
```
