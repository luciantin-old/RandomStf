```python
import cv2
import numpy as np

image = cv2.imread('./images/flower1.jpg')

height, width = image.shape[:2]

quarterHeight, quarterWidth = height/4 , width/4

#T is the translation matrix
T = np.float32( [ [1,0,quarterWidth],[0,1,quarterHeight] ] )

imgTranslation = cv2.warpAffine(image,T, (width,height))

cv2.imshow('',imgTranslation)

# cv2.imshow('',image)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


square = np.zeros((300,300), np.uint8)
cv2.rectangle(square, (50,50), (250,250), 255, -2)

ellipse = np.zeros((300,300),np.uint8)
cv2.ellipse(ellipse, (150,150), (150,150), 30, 0, 180, 255, -1)

cv2.waitKey()
cv2.destroyAllWindows()

AND = cv2.bitwise_and(square,ellipse)
cv2.imshow('',AND)

cv2.waitKey()
cv2.destroyAllWindows()

OR = cv2.bitwise_or(square,ellipse)
cv2.imshow('',OR)

cv2.waitKey()
cv2.destroyAllWindows()

XOR = cv2.bitwise_xor(square,ellipse)
cv2.imshow('',XOR)

cv2.waitKey()
cv2.destroyAllWindows()

NOT = cv2.bitwise_not(ellipse)
cv2.imshow('',NOT)

cv2.waitKey()
cv2.destroyAllWindows()


```

```python
import cv2 
import numpy as np

image = cv2.imread('./images/daisy.jpg',0)

kernel = np.ones((5,5), np.uint8)

cv2.imshow('',image)
cv2.waitKey()

erosion = cv2.erode(image,kernel,iterations = 1)
cv2.imshow('',erosion)
cv2.waitKey()

dilation = cv2.dilate(image,kernel, iterations = 1)
cv2.imshow('',dilation)
cv2.waitKey()

opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
cv2.imshow('',opening)
cv2.waitKey()

closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
cv2.imshow('',closing)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np

image = cv2.imread('./images/daisy.jpg',0)

height, width = image.shape

image = cv2.bilateralFilter(image,9,75,75)
image = cv2.erode(image, np.ones((5,5),np.uint8),iterations = 1)

cv2.imshow('', image)
cv2.waitKey()

sobelX = cv2.Sobel(image, cv2.CV_64F, 0,1, ksize=5)
sobelY = cv2.Sobel(image, cv2.CV_64F, 1,0, ksize=5)

cv2.imshow('', sobelX)
cv2.waitKey()
cv2.imshow('', sobelY)
cv2.waitKey()

sobel_OR = cv2.bitwise_or(sobelX,sobelY)

cv2.imshow('sobel',sobel_OR)
cv2.waitKey()

laplacian = cv2.Laplacian(image, cv2.CV_64F)
cv2.imshow('laplace',laplacian)
cv2.waitKey()



canny = cv2.Canny(image, 20,70)
cv2.imshow('canny',canny)
cv2.waitKey()


cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np

def sketch(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    cannyEdge = cv2.Canny(blur,10,60)
    ret,mask = cv2.threshold(cannyEdge, 70, 255, cv2.THRESH_BINARY)
    mask = cv2.dilate(mask, np.ones((5,5),np.uint8), iterations = 1)
    mask = cv2.bitwise_not(mask)
    return mask

cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    cv2.imshow('dap',sketch(frame))
    if cv2.waitKey(1) == 13:
        break

cv2.destroyAllWindows()
cap.release()
```
